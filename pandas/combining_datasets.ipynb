```python
def make_df(cols,ind):
    """Quickly nake a DataFrame"""
    data={c:[str(c)+str(i) for i in ind] 
          for c in cols}
    return pd.DataFrame(data,ind)
make_df('ABC',range(3))
```

```python
fd1=make_df('AB',[1,2])
fd2=make_df('AB',[3,4])
print(fd1)
print(fd2)
```

```python
pd.concat([fd1,fd2])
```

```python
fd3=make_df('AB',[0,1])
fd4=make_df('AB',[2,3])
print(fd3)
print(fd4)
pd.concat([fd3,fd4],axis=1)
```

```python
abbrevs=pd.read_csv('../PythonDataScienceHandbook/notebooks/data/state-abbrevs.csv')
abbrevs.head()
```

```python
merged=pd.merge(pop,abbrevs,left_on='state/region',right_on='abbreviation',how='outer')
merged.head()
```

```python
merged.count()
```

```python
merged=merged.drop(labels='abbreviation',axis=1)
```
