```python
mean=[0,0]
cov=[[1,1],[1,2]]
x,y=np.random.multivariate_normal(mean,cov,10000).T
```

```python
plt.hist2d(x,y,bins=50,cmap='viridis')
cb=plt.colorbar()
cb.set_label('count in bin')
```

```python
plt.hist(x,alpha=0.3,bins=50)
plt.hist(y,alpha=0.3,bins=50)
```

## Hexagonal binnings

```python
plt.hexbin(x,y,gridsize=30,cmap='Blues')
cb=plt.colorbar(label='count in bin')
```

```python
plt.hexbin(x,y,gridsize=20,cmap='Greens')
cb=plt.colorbar(label='count in bin')
```

### kernel density estimation

```python
from scipy.stats import gaussian_kde
```

```python
data=np.vstack([x,y])
```

```python
kde=gaussian_kde(data)
```

```python
xgrid=np.linspace(-3.5,3.5,400)
ygrid=np.linspace(-6,6,40)
Xgrid,Ygrid=np.meshgrid(xgrid,ygrid)
```

```python
Z=kde.evaluate(np.vstack([Xgrid.ravel(),Ygrid.ravel()]))
```

```python
plt.imshow(Z.reshape(Xgrid.shape),origin='lower',aspect='auto',extent=[-3.5,3.5,-6,6],cmap='Blues')
cb=plt.colorbar(label='density')
```
