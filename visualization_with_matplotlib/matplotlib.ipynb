# Gaussian Process Regressor

```python
from sklearn.gaussian_process import GaussianProcessRegressor
```

```python
GaussianProcessRegressor?
```

```python
gp=GaussianProcessRegressor()
```

```python
#x=np.linspace(0,100,100)
x=[1,3,5,6,8]
y=x*np.sin(x)
plt.plot(x,y)
```

```python
x=np.array([1,3,5,6,8])
y=x*np.sin(x)
```

```python
gp.fit(x[:,np.newaxis],y)
```

```python
xfit=np.linspace(0,10,1000)
yfit, MSE=gp.predict(xfit[:,np.newaxis],return_std=True)
ycalc=xfit*np.sin(xfit)
```

```python
dyfit=2*np.sqrt(MSE)
```

```python
#visualize
plt.plot(x,y,'or')
plt.plot(xfit,yfit,'-',color='gray')
```

```python
plt.plot(x,y,'or')
plt.plot(xfit,yfit,'-',color='gray')
plt.scatter(xfit,ycalc,c='m',s=10,alpha=0.2)
plt.fill_between(xfit,yfit-dyfit,yfit+dyfit,color='gray',alpha=0.5)
```

```python
x=np.array([1,3,5,6,8])
model=lambda x:x*np.sin(x)
ydata=model(x)
plt.plot(x,y)
plt.plot(x,ydata,alpha=0.2,linewidth=10)
```
