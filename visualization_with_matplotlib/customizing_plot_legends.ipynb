```python
plt.scatter(lon,lat,label=None, c=np.log10(population),cmap='viridis',s=area,linewidths=0,alpha=0.5)
plt.colorbar()
plt.clim(3,7)
```

## multiple legends

```python
import seaborn as sns
sns.set()
```

```python
fig ,ax =plt.subplots()
lines=[]
styles=['-','--','-.',':']
#colors=['maroon','khaki','olive','teal']
colors=['turquoise','cornflowerblue','thistle','indigo']
x=np.linspace(0,10,1000)

for i in range(len(styles)):
    lines+=ax.plot(x,np.sin(x-i*np.pi/2),styles[i],color=colors[i])
    
ax.axis('equal')

ax.legend(lines[:2],['A','B'],loc='upper right',frameon=False)

####
from matplotlib.legend import Legend
leg=Legend(ax, lines[2:],['C','D'],loc='lower right',frameon=False)
ax.add_artist(leg)
```
