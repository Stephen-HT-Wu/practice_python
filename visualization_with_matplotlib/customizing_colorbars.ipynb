### choosing the colormap

```python
from matplotlib.colors import LinearSegmentedColormap
```

```python
cmap=plt.cm.get_cmap(plt.cm.jet)
```

```python
a=cmap(np.arange(cmap.N))
print(a.shape)
labels=['B','G','R']
colors=[0,128,255]
for i in range(3):
    plt.plot(a[:,i],label=labels[i],c=cmap(colors[i]))
    
plt.legend()
```

```python
x=np.arange(256)
I=x+(x[:,np.newaxis])
I=I/2
print(I.shape)
plt.imshow(I,cmap='jet',origin='lower')
plt.colorbar()
```

```python
plt.contourf(I,20,cmap=plt.cm.jet,)
plt.colorbar()
```

```python
colors=cmap(np.arange(cmap.N))
```

```python
RGB_weight=[0.299,0.587,0.114]
luminance=np.sqrt(np.dot(colors[:,:3]**2,RGB_weight))
```

```python
luminance.shape
```

```python
n=luminance[:,np.newaxis]
```

```python
plt.plot(n,label='luminance',c='chartreuse',linewidth=3,alpha=0.3)
plt.plot(colors[:,0],':',label='B',c='b')
plt.plot(colors[:,1],':',label='G',c='g')
plt.plot(colors[:,2],':',label='R',c='r')
plt.legend()
```

```python
new_colors=np.empty([256,4])
new_colors[:,:3]=luminance[:,np.newaxis]
new_colors[:,3]=colors[:,3]
```
