```python
x=10*rng.rand(50)
y=2*x-1 + rng.randn(50)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression(fit_intercept=True)
```

```python
model
```

```python
X=x[:,np.newaxis]
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
xfit=np.linspace(-1,11) # default num=50
```

```python
Xfit=xfit[:,np.newaxis]
yfit=model.predict(Xfit)
```

```python
plt.scatter(x,y,c='orange',alpha=0.3)
plt.plot(xfit,yfit)
```
